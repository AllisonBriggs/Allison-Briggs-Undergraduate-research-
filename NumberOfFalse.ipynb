```python
import pandas as pd
from scipy.stats import chi2_contingency
#excel_file_path = 'thinkaloud_action_level_data.csv'
df  = pd.read_csv("thinkaloud_action_level_data.csv")
#print(df.columns)
#correctness_false_df = df[df['Correctness'] == "FALSE"]
df.head()


```

```python
df.dtypes
```

```python
df['Correctness'].value_counts()
```

```python
df1= df.query('Correctness == False')
df1.head()
```

```python
df2 = df1.groupby(['Problem', 'ProblemStep', 'Answer.Text'], as_index=False).agg(Answer_Counts = ('Answer.Text', 'count'))
```

```python
df2.columns
```

```python
ruledf  = pd.read_csv("rule_types.csv")
ruledf.head()
```

```python
df3 = pd.merge(df2, ruledf, how="left", left_on=["Problem", "ProblemStep"], right_on=["Problem", "ProblemStep"])
df3
# Answer.text == unique count
```

```python
df4 = pd.merge(df1, ruledf, how="left", left_on=["Problem", "ProblemStep"], right_on=["Problem", "ProblemStep"])
df4
```

```python
contigency = pd.crosstab(index = df4["Answer.Text"], columns = df4["Rule_type"])
contigency
```

```python
chi2_contingency(contigency)
```
